# Comparison of results and data analysis

We perform a comparison of the results obtained from the systems under inspection

```python
%autosave 0
import matplotlib.pyplot as plt
import numpy as np
import Routine as R
import json
virtOrbitalsPath = 'virtual-orbitals/'
```

```python
systems = ['N2','CO','C60']
```

Read engVirt and S_nivrt_Ez

```python
engVirt = []
for s in systems:
    fileName = s + '/' + virtOrbitalsPath + 'engVirt.txt'
    print fileName
    with open(fileName) as f: 
        e = json.load(f)
    engVirt.append(e)
    
S_nvirt_Ez = []
for s in systems:
    fileName = s + '/' + virtOrbitalsPath + 'S_nvirt_Ez.txt'
    print fileName
    with open(fileName) as f: 
        s = json.load(f)
    S_nvirt_Ez.append(s)
```

Plot the maximum energy in function of the number of virtual orbitals

```python
for i,s in enumerate(systems): 
    plt.plot(engVirt[i],label=s)
plt.legend()
plt.title('Maximum Energy vs. number of virtual orbitals',fontsize=12)
plt.show()
```

Plot S_nvirt_Ez in function of the energy of the virtual orbitals

```python
%matplotlib inline
##%matplotlib auto 
for i,s in enumerate(systems): 
    plt.semilogy(engVirt[i],S_nvirt_Ez[i],label=s)
plt.xlabel('energy(eV)')
plt.legend()
plt.show()
```

Plot S_nvirt_Ez in function of the number of the virtual orbitals

```python
%matplotlib inline
for i,s in enumerate(systems): 
    plt.semilogy(S_nvirt_Ez[i],label=s)
plt.legend()
plt.show()
```

Comments:

* It seems that the maximum energy is the relevant parameter to achieve a certain level of completeness. So, for istance since the $C60$ grows slower in energy it $S_{nvirt}$ value is worst than the others. This fact suggests that it could be interesting to remove the virtual states that do not produce an increasing of the maximum energy.
